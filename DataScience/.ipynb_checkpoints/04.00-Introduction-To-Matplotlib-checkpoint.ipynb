````python
import matplotlib as mpl
import matplotlib.pyplot as plt
````

### Setting Styles

````python
plt.style.use('classic')
````

### ``show()`` or No ``show()``? How to Display Your Plots

#### Plotting from a script

``plt.show()`` (once, at end of script) starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.

```python
# ------- file: myplot.py ------
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

#### Plotting from an IPython shell

```ipython
In [1]: %matplotlib
Using matplotlib backend: TkAgg

In [2]: import matplotlib.pyplot as plt
```

Any ``plt`` plot command will cause a figure window to open: to force an update, use ``plt.draw()``.
No ``plt.show()`` required.

#### Plotting from an IPython notebook

- ``%matplotlib notebook`` will lead to *interactive* plots embedded within the notebook
- ``%matplotlib inline`` will lead to *static* images of your plot embedded in the notebook

````python
%matplotlib inline
````

After running this command (it needs to be done only once per kernel/session), any cell within the notebook that creates a plot will embed a PNG image of the resulting graphic:

````python
import numpy as np
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
````

    ## Two Interfaces for the Price of One

### MATLAB-style Interface: stateful

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

But, once the second panel is created, how can we go back and add something to the first?
This is possible within the MATLAB-style interface, but a bit clunky.
Fortunately, there is a better way.

#### Object-oriented interface

Plotting functions are *methods* of explicit ``Figure`` and ``Axes`` objects.

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````
